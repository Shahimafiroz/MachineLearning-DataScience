```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
doge = pd.read_csv('DOGE-USD.csv')
```

```python
X = doge.iloc[ : , 1: ]
Y = doge.iloc[ : , 1: ]
doge.head()
```

```python
plt.figure(figsize=(20, 6))  # Set the figure size
plt.plot(doge['Date'], doge['Open'], label='Open', color='blue')
plt.plot(doge['Date'], doge['Close'], label='Close', color='green')
plt.plot(doge['Date'], doge['High'], label='High', color='pink')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Cryptocurrency Price Trends Over Time')
plt.legend()  # Show the legend
```

```python
doge.isnull().values.any()
```

```python
from sklearn.model_selection import train_test_split as TrTes
from sklearn.linear_model import LinearRegression as LR
```

```python
# Select your independent variable and dependent variable
x = doge[['Open']]  # Independent variable (features)
y = doge['Close']  # Dependent variable (target)

```

```python
x_train, x_test ,y_train ,y_test = TrTes(x , y , test_size=0.2 , random_state=0)
model = LR()
model.fit(x_train , y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
plt.scatter(x_test ,y_test , label='Actual' , color = 'blue')
plt.plot(x_test , y_predict , color='red', linewidth=2, label='Predicted' )
plt.xlabel('Open Price')
plt.ylabel('Close Price')
plt.title('Linear Regression for Price Prediction')
plt.legend()
plt.show()
```
